## 2.2 Recurrent Neural Network (RNN)

## 2.2.1 Why do we need RNN? What is the unique value?

Convolutional neural networks - CNNs and common algorithms are mostly one-to-one correspondence between input and output, i.e., one input gets one output. There is no connection between the different inputs. The RNN is derived to solve this problem.

Original idea was introduced by Rumelhart, David E et al. (1985)

Source: https://apps.dtic.mil/dtic/tr/fulltext/u2/a164453.pdf


## 2.2.2 The basic principle of RNN

The biggest difference between RNN and traditional neural networks is that each time, the output of the previous one is brought to the next hidden layer and trained together.

## 2.3 Long Short-Term Memory cell (LSTM)

Long-Short Term Memory (LSTM) artificial neural network paper was first published in 1997. Due to the unique design structure, LSTM is suitable for processing and predicting important events with very long intervals and delays in time series.

Source: https://ieeexplore.ieee.org/abstract/document/6795963





The forgetting gate, the input gate (update gate in the figure) and the output gate are each a small single-layer neural network with a sigmoid activation function. Since the sigmoid takes values in the range \(0, 1\), it is effectively used to determine whether to retain or "forget" information (multiplying by a value close to 1 for retention and close to 0 for forgetting), providing us with the ability to selectively transmit information. ### 2.3.1 How transformer or self-attention algorithmus works?

We generate three corresponding vectors for the input word vector \(X\): Query, Key and Value. note that these three vectors are much smaller than the vector \(X\) (the length of \(X\) in the paper is 512 and the length of the three vectors is 64, which is only an architectural choice based on the fact that the Multi-Head Attention in the paper has 8 Self-Attention modules, and the output of the 8 Self-Attentions are to be spliced to recover them into a vector of length 512), this one part is operated independently for each word.

The score is divided by the square root of the vector dimension (64) (to ensure that the score is in a small range, otherwise the softmax result is non-zero, i.e., 1), and then it is Softmaxed (to normalize the scores of all words so that all words are positive and sum to 1). This way for each word will get the score of all words contributing to the encoding of the word, of course the current word will get the maximum score, but will also pay attention to the size of the contribution of other words.

For the obtained Softmax score, we multiply it by each corresponding Value vector (weakening the effect of words with lower softmax scores, somewhat similar to the idea of the previous sigmoid gate function)
The output of Self-Attention for the current word "Thinking" is obtained by summing all the resulting weighted vectors.

- self attention in one head

```python
# define image extract function
import cv2
import numpy as np

def get_frames(filename, n_frames=1):
    frames = []
    v_cap = cv2.VideoCapture(filename)
    v_len = int(v_cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_list = np.linspace(0, v_len - 1, n_frames + 1, dtype=np.int16)

    for fn in range(v_len):
        success, frame = v_cap.read()
        if success is False:
            continue
        if (fn in frame_list):
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frames.append(frame)
    v_cap.release()
    return frames, v_len


def store_frames(frames, path2store):
    for ii, frame in enumerate(frames):
        frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
        path2img = os.path.join(path2store, "frame" + str(ii) + ".jpg")
        cv2.imwrite(path2img, frame)
```

```python
# extract label information from video names
import os

path2data = 'assets/hmdb51'
sub_folder_jpg = "hmdb51_jpg"
path2ajpgs = os.path.join(path2data, sub_folder_jpg)

def get_vids(path2ajpgs):
    listOfCats = os.listdir(path2ajpgs)
    ids = []
    labels = []
    for catg in listOfCats:
        path2catg = os.path.join(path2ajpgs, catg)
        listOfSubCats = os.listdir(path2catg)
        path2subCats = [os.path.join(path2catg, los) for los in listOfSubCats]
        ids.extend(path2subCats)
        labels.extend([catg] * len(listOfSubCats))
    return ids, labels, listOfCats
```

```python
# make dataset from extracted images 
from sklearn.model_selection import StratifiedShuffleSplit

all_vids, all_labels, catgs = get_vids(path2ajpgs)
print(len(all_vids), len(all_labels), len(catgs))

labels_dict = {}
ind = 0
for uc in catgs:
    labels_dict[uc] = ind
    ind += 1

print(labels_dict, 'labels_dict')

num_classes = 5
unique_ids = [id_ for id_, label in zip(all_vids, all_labels) if labels_dict[label] < num_classes]
unique_labels = [label for id_, label in zip(all_vids, all_labels) if labels_dict[label] < num_classes]
print('Video numbers:', len(unique_ids), 'Label numbers:', len(unique_labels))

sss = StratifiedShuffleSplit(n_splits=2, test_size=0.1, random_state=0)
train_indx, test_indx = next(sss.split(unique_ids, unique_labels))

train_ids = [unique_ids[ind] for ind in train_indx]
train_labels = [unique_labels[ind] for ind in train_indx]
print('Training videos:', len(train_ids), 'Training labels:', len(train_labels))

test_ids = [unique_ids[ind] for ind in test_indx]
test_labels = [unique_labels[ind] for ind in test_indx]
print('Test videos:', len(test_ids), 'Test labels:', len(test_labels))
```

```python
# define video VideoDataset
import glob
from PIL import Image
import random
import torch
from torch.utils.data import Dataset
import numpy as np

timesteps =16

# set seed for random number
np.random.seed(2020)
random.seed(2020)
torch.manual_seed(2020)

class VideoDataset(Dataset):
    def __init__(self, ids, labels, transform):
        self.transform = transform
        self.ids = ids
        self.labels = labels

    def __len__(self):
        return len(self.ids)

    def __getitem__(self, idx):
        path2imgs = glob.glob(self.ids[idx] + "/*.jpg")
        path2imgs = path2imgs[:timesteps]
        label = labels_dict[self.labels[idx]]
        frames = []
        for p2i in path2imgs:
            frame = Image.open(p2i)
            frames.append(frame)

        seed = np.random.randint(1e9)
        frames_tr = []
        for frame in frames:
            random.seed(seed)
            np.random.seed(seed)
            frame = self.transform(frame)
            frames_tr.append(frame)
        if len(frames_tr) > 0:
            frames_tr = torch.stack(frames_tr)
        return frames_tr, label
```

```python
# Dataloader
from torch.utils.data import DataLoader

batch_size = 4

def collate_fn_r3d_18(batch):
    imgs_batch, label_batch = list(zip(*batch))
    imgs_batch = [imgs for imgs in imgs_batch if len(imgs) > 0]
    label_batch = [torch.tensor(l) for l, imgs in zip(label_batch, imgs_batch) if len(imgs) > 0]
    imgs_tensor = torch.stack(imgs_batch)
    imgs_tensor = torch.transpose(imgs_tensor, 2, 1)
    labels_tensor = torch.stack(label_batch)
    return imgs_tensor, labels_tensor


def collate_fn_rnn(batch):
    imgs_batch, label_batch = list(zip(*batch))
    imgs_batch = [imgs for imgs in imgs_batch if len(imgs) > 0]
    label_batch = [torch.tensor(l) for l, imgs in zip(label_batch, imgs_batch) if len(imgs) > 0]
    imgs_tensor = torch.stack(imgs_batch)
    labels_tensor = torch.stack(label_batch)
    return imgs_tensor, labels_tensor



if model_type == "rnn" or model_type == "lstm" or model_type == 'transformer':
    train_dl = DataLoader(train_ds, batch_size=batch_size,
                          shuffle=True, collate_fn=collate_fn_rnn)
    test_dl = DataLoader(test_ds, batch_size=2 * batch_size,
                         shuffle=False, collate_fn=collate_fn_rnn)

else:
    train_dl = DataLoader(train_ds, batch_size=batch_size,
                          shuffle=True, collate_fn=collate_fn_r3d_18)
    test_dl = DataLoader(test_ds, batch_size=2 * batch_size,
                         shuffle=False, collate_fn=collate_fn_r3d_18)
```

```python
# setting rnn model

from torch import nn

class Resnt18Rnn(nn.Module):
    def __init__(self, params_model):
        super(Resnt18Rnn, self).__init__()
        num_classes = params_model["num_classes"]
        dr_rate = params_model["dr_rate"]
        pretrained = params_model["pretrained"]
        rnn_hidden_size = params_model["rnn_hidden_size"]
        rnn_num_layers = params_model["rnn_num_layers"]
        model_type = params_model["model_type"]

        baseModel = models.resnet18(pretrained=pretrained)
        num_features = baseModel.fc.in_features
        baseModel.fc = Identity()
        self.model_type = model_type
        self.baseModel = baseModel
        self.dropout = nn.Dropout(dr_rate)

        # RNN layer
        self.rnn = nn.RNN(num_features, rnn_hidden_size, rnn_num_layers)
        
        # LSTM cell
        self.lstm = nn.LSTM(num_features, rnn_hidden_size, rnn_num_layers)

        # Transformer encoder
        self.encoder_layer = nn.TransformerEncoderLayer(d_model=num_features, nhead=8)
        self.transformer_encoder = nn.TransformerEncoder(self.encoder_layer, num_layers=6)

        self.fc1 = nn.Linear(rnn_hidden_size, num_classes)
        self.fc_trans = nn.Linear(num_features, num_classes)

    def forward(self, x):
        b_z, ts, c, h, w = x.shape
        ii = 0
        y = self.baseModel((x[:, ii]))

        if self.model_type == "rnn":
            output, hn = self.rnn(y.unsqueeze(1))
            for ii in range(1, ts):
                y = self.baseModel((x[:, ii]))
                out, hn = self.rnn(y.unsqueeze(1), hn)

        elif self.model_type == "lstm":
            output, (hn, cn) = self.lstm(y.unsqueeze(1))
            for ii in range(1, ts):
                y = self.baseModel((x[:, ii]))
                out, (hn, cn) = self.lstm(y.unsqueeze(1), (hn, cn))

        elif self.model_type == "transformer":
            feature = []
            for ii in range(0, ts):
                y = self.baseModel((x[:, ii]))
                y_unsqueeze = y.unsqueeze(1)
                if ii == 0:
                    input_feature = y_unsqueeze
                else:
                    input_feature = torch.cat((input_feature, y_unsqueeze), 1)

            out = self.transformer_encoder(input_feature)

        out = self.dropout(out[:, -1])

        if self.model_type == "transformer":
            out = self.fc_trans(out)
        else:
            out = self.fc1(out)

        return out
```

```python
import copy
from tqdm import tqdm

def get_lr(opt):
    for param_group in opt.param_groups:
        return param_group['lr']
    
    
def loss_epoch(model, loss_func, dataset_dl, sanity_check=False, opt=None):
    running_loss = 0.0
    running_metric = 0.0
    len_data = len(dataset_dl.dataset)
    # for xb, yb in tqdm_notebook(dataset_dl):
    for xb, yb in tqdm(dataset_dl):
        xb = xb.to(device)
        yb = yb.to(device)
        output = model(xb)
        loss_b, metric_b = loss_batch(loss_func, output, yb, opt)
        running_loss += loss_b

        if metric_b is not None:
            running_metric += metric_b
        if sanity_check is True:
            break
    loss = running_loss / float(len_data)
    metric = running_metric / float(len_data)
    return loss, metric

        
def loss_batch(loss_func, output, target, opt=None):
    loss = loss_func(output, target)
    with torch.no_grad():
        metric_b = metrics_batch(output, target)
    if opt is not None:
        opt.zero_grad()
        loss.backward()
        opt.step()
    return loss.item(), metric_b


def metrics_batch(output, target):
    pred = output.argmax(dim=1, keepdim=True)
    corrects = pred.eq(target.view_as(pred)).sum().item()
    return corrects
```

```python
# setting training and validation process


def train_val(model, params):
    num_epochs = params["num_epochs"]
    loss_func = params["loss_func"]
    opt = params["optimizer"]
    train_dl = params["train_dl"]
    val_dl = params["val_dl"]
    sanity_check = params["sanity_check"]
    lr_scheduler = params["lr_scheduler"]
    path2weights = params["path2weights"]

    loss_history = {
        "train": [],
        "val": [],
    }

    metric_history = {
        "train": [],
        "val": [],
    }

    best_model_wts = copy.deepcopy(model.state_dict())
    best_loss = float('inf')

    for epoch in range(num_epochs):
        current_lr = get_lr(opt)
        print('Epoch {}/{}, current lr={}'.format(epoch, num_epochs - 1, current_lr))
        model.train()
        train_loss, train_metric = loss_epoch(model, loss_func, train_dl, sanity_check, opt)
        loss_history["train"].append(train_loss)
        metric_history["train"].append(train_metric)
        model.eval()
        with torch.no_grad():
            val_loss, val_metric = loss_epoch(model, loss_func, val_dl, sanity_check)
        if val_loss < best_loss:
            best_loss = val_loss
            best_model_wts = copy.deepcopy(model.state_dict())
            torch.save(model.state_dict(), path2weights)
            print("Copied best model weights!")

        loss_history["val"].append(val_loss)
        metric_history["val"].append(val_metric)

        lr_scheduler.step(val_loss)
        if current_lr != get_lr(opt):
            print("Loading best model weights!")
            model.load_state_dict(best_model_wts)

        print("train loss: %.6f, dev loss: %.6f, accuracy: %.2f" % (train_loss, val_loss, 100 * val_metric))
        print("-" * 10)
    model.load_state_dict(best_model_wts)

    return model, loss_history, metric_history

# start traning

model, loss_hist, metric_hist = train_val(model, params_train)
```

```python
# make a dataloder for HOIsim


class HoisimDataloder(Dataset):

    '''
    Make a Hoisim dataset and feed to Dataloder afterwards
    __getitem__ :   To get data dependng on the index you give
    __len__:        Get the length of the dataset
    '''

    def __init__(self, data_dir, transform=None):
        self.data_labels, self.data_features = self.make_dataset(data_dir)
        self.transform = transform

    def __getitem__(self, index):
        return self.data_labels[index], self.data_features[index]

    def __len__(self):
        return len(self.data_features)


    @staticmethod
    def make_dataset(data_dir):

        ACTIONS_LIST = ['GoTo', 'Open', 'Close', 'Pickup', 'Place', 'Stack', 'Animate(PastaInWater: 1)',
                            'Animate(SwirlSauce)', 'Animate(SprinkleOil)', 'Animate(TossTomatoIntoPan)',
                            'Animate(CutTomato)',
                            'Animate(FillWater)', 'Animate(StrainPasta)', 'Animate(SauceToPot)', 'Animate(EatPasta)',
                            'Animate(PastaToPlate)', 'Animate(PastaInWater)', 'Animate(ShakeCereal)',
                            'Animate(PourMilk)',
                            'Animate(StirSpoon)', 'Animate(EatWithSpoon)', 'UNKNOWN']

        OBJECTS_LIST = ['BigPotWithLid', 'Bowl', 'ButterKnife', 'Lid', 'Pan', 'Spoon', 'UNKNOWN', 'Wooden Fork 1',
                        'ball_l', 'ball_l_end', 'ball_r', 'ball_r_end', 'calf_l', 'calf_r', 'chukies_02', 'clavicle_l',
                        'clavicle_r', 'foot_l', 'foot_r', 'hand_l', 'hand_r', 'head', 'head_end', 'lowerarm_l',
                        'lowerarm_r',
                        'milk', 'neck_01', 'oil', 'pasta', 'pelvis', 'spine_01', 'spine_02', 'spine_03', 'thigh_l',
                        'thigh_r',
                        'tomato', 'upperarm_l', 'upperarm_r', 'Fork', 'Plate_1', 'CuttingBoard_1', 'kitchen',
                        'Table', 'KitchenCounter', 'Fridge Prefab', 'Stove',
                        'Cupboard_3R_T', 'Sink', 'BottomDrawerSectionRight_5']

        len_object = len(OBJECTS_LIST)

        target_action = []
        object_features = []
        st_object_features = []
        frame_obejct_features = []
        tmp_object_features = 0
        tmp_seq = 0
        frame_count = 0

        directory = os.path.join(data_dir)
        for root, dirs, files in os.walk(directory):
            for file in tqdm(files):
                if file.endswith(".csv"):
                    # f = open(root + '/' + file, 'r')
                    csv_data = pd.read_csv(root + '/' + file)
                    for num, rows in enumerate(csv_data.iterrows()):
                        curr_seq = rows[1]['seq']
                        if curr_seq != tmp_seq:
                            action = rows[1]['action']
                            action_class = ACTIONS_LIST.index(action)
                            target_action.append(action_class)
                            tmp_seq = curr_seq

                            # object_features.append(frame_obejct_features)
                            if frame_count != 0:
                                st_graph = np.array(frame_obejct_features) - np.array(tmp_object_features)
                                tmp_object_features = frame_obejct_features
                                flatten_st_graph = [item for sublist in st_graph for item in sublist]
                            # st_graph = frame_obejct_features - tmp_object_features
                                st_object_features.append(flatten_st_graph)

                            frame_obejct_features = []
                            frame_count += 1

                        object_matrix = rows[1][7:23].values
                        # object_matrix = rows[1][7:23].values.reshape((4, 4))
                        frame_obejct_features.append(object_matrix)



        return  st_object_features, target_action[1:]
```

```python
# collate function that transfer data to Tensor type

def feature_to_tensor(feature):
    return torch.tensor(feature)

def collate_fn(batch):
    '''
    collate function here is used to convert data to Tensor
    so that we can feed them to gpu device later
    '''
    st_graph_batch, label_batch = list(zip(*batch))
    st_graph_batch = torch.unsqueeze(feature_to_tensor(st_graph_batch), dim=1)
    st_graph_batch = [imgs for imgs in st_graph_batch if len(imgs) > 0]
    label_batch = [torch.tensor(l) for l, feature in zip(label_batch, st_graph_batch) if len(feature) > 0]
    st_graph_batch_tensor = torch.stack(st_graph_batch)
    labels_tensor = torch.stack(label_batch)
    return st_graph_batch_tensor, labels_tensor

```

```python
hoisim_dataset = HoisimDataloder(data_path)
print(hoisim_dataset)
hoisim_loader = DataLoader(hoisim_dataset,
                           batch_size=4,
                           shuffle=False,
                           collate_fn=collate_fn)
```

```python
# setting sequential models

class squential_models(nn.Module):
    def __init__(self, params_model):
        super(squential_models, self).__init__()

        num_classes = params_model["num_classes"]
        dr_rate = params_model["dr_rate"]
        pretrained = params_model["pretrained"]
        rnn_hidden_size = params_model["rnn_hidden_size"]
        rnn_num_layers = params_model["rnn_num_layers"]
        model_type = params_model["model_type"]
        num_features = params_model["num_features"]

        self.model_type = model_type

        # Dropout layer
        self.dropout = nn.Dropout(dr_rate)

        # RNN layer
        self.rnn = nn.RNN(num_features, rnn_hidden_size, rnn_num_layers)
        # LSTM cell
        self.lstm = nn.LSTM(num_features, rnn_hidden_size, rnn_num_layers)

        # Transformer encoder
        self.encoder_layer = nn.TransformerEncoderLayer(d_model=num_features, nhead=8)
        self.transformer_encoder = nn.TransformerEncoder(self.encoder_layer, num_layers=6)

        # Fully connected layer
        self.fc1 = nn.Linear(rnn_hidden_size, num_classes)
        # Fully connected layer for transformer
        self.fc_trans = nn.Linear(num_features, num_classes)

    def forward(self, x):
        if self.model_type == "rnn":
            x, hn = self.rnn(x)

        elif self.model_type == "lstm":
            x, hn = self.lstm(x)

        elif self.model_type == "transformer":
            x = self.transformer_encoder(x)

        x = self.dropout(x[:, -1])

        if self.model_type == "transformer":
            out = self.fc_trans(x)
        else:
            out = self.fc1(x)

        return out
```

```python
# define class number and feature dimentions

num_classes = 49
num_features = len(hoisim_dataset[0][0])
```

```python
# setting model parameters

params_model = {
    "model_type": model_type,
    "num_classes": num_classes,
    "dr_rate": 0.1,
    "pretrained": True,
    "rnn_num_layers": 1,
    "rnn_hidden_size": 100,
    "num_features": num_features}
```

```python
# initialize model

model = squential_models(params_model)
```

```python
# Model Training

path2weights = "assets/hoisim/model/weights.pt"
torch.save(model.state_dict(), path2weights)
```

```python
# feed data to your gpu
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
# setting your loss function, optimizer and learning rate

loss_func = nn.CrossEntropyLoss(reduction="sum")

if model_type == 'transformer':
    lr = 'you need to fill it here'
else:
    lr = 1e-4

opt = optim.Adam(model.parameters(), lr=lr)
lr_scheduler = ReduceLROnPlateau(opt, mode='min', factor=0.5, patience=5, verbose=1)
```

```python
# seting your traning parameters

params_train = {
    "num_epochs": 10,
    "optimizer": opt,
    "loss_func": loss_func,
    "train_dl": hoisim_loader,
    "val_dl": hoisim_loader,
    "sanity_check": False,
    "lr_scheduler": lr_scheduler,
    "path2weights": "assets/hoisim/model/weights_"+model_type+".pt",
    }
```

```python
# setting training update functions

def get_lr(opt):
    for param_group in opt.param_groups:
        return param_group['lr']

def metrics_batch(output, target):
    pred = output.argmax(dim=1, keepdim=True)
    corrects = pred.eq(target.view_as(pred)).sum().item()
    return corrects


def loss_batch(loss_func, output, target, opt=None):
    loss = loss_func(output, target)
    with torch.no_grad():
        metric_b = metrics_batch(output, target)
    if opt is not None:
        opt.zero_grad()
        loss.backward()
        opt.step()
    return loss.item(), metric_b


def loss_epoch(model, loss_func, dataset_dl, sanity_check=False, opt=None):
    running_loss = 0.0
    running_metric = 0.0
    len_data = len(dataset_dl.dataset)
    # for xb, yb in tqdm_notebook(dataset_dl):
    for xb, yb in tqdm(dataset_dl):
        xb = xb.to(device)
        yb = yb.to(device)
        output = model(xb)
        loss_b, metric_b = loss_batch(loss_func, output, yb, opt)
        running_loss += loss_b

        if metric_b is not None:
            running_metric += metric_b
        if sanity_check is True:
            break
    loss = running_loss / float(len_data)
    metric = running_metric / float(len_data)
    return loss, metric
```

```python

def train_val(model, params):
    num_epochs = params["num_epochs"]
    loss_func = params["loss_func"]
    opt = params["optimizer"]
    train_dl = params["train_dl"]
    val_dl = params["val_dl"]
    sanity_check = params["sanity_check"]
    lr_scheduler = params["lr_scheduler"]
    path2weights = params["path2weights"]

    loss_history = {
        "train": [],
        "val": [],
    }

    metric_history = {
        "train": [],
        "val": [],
    }

    best_model_wts = copy.deepcopy(model.state_dict())
    best_loss = float('inf')

    for epoch in range(num_epochs):
        current_lr = get_lr(opt)
        print('Epoch {}/{}, current lr={}'.format(epoch, num_epochs - 1, current_lr))
        model.train()
        train_loss, train_metric = loss_epoch(model, loss_func, train_dl, sanity_check, opt)
        loss_history["train"].append(train_loss)
        metric_history["train"].append(train_metric)
        model.eval()
        with torch.no_grad():
            val_loss, val_metric = loss_epoch(model, loss_func, val_dl, sanity_check)
        if val_loss < best_loss:
            best_loss = val_loss
            best_model_wts = copy.deepcopy(model.state_dict())
            torch.save(model.state_dict(), path2weights)
            print("Copied best model weights!")

        loss_history["val"].append(val_loss)
        metric_history["val"].append(val_metric)

        lr_scheduler.step(val_loss)
        if current_lr != get_lr(opt):
            print("Loading best model weights!")
            model.load_state_dict(best_model_wts)

        print("train loss: %.6f, dev loss: %.6f, accuracy: %.2f" % (train_loss, val_loss, 100 * val_metric))
        print("-" * 10)
    model.load_state_dict(best_model_wts)

    return model, loss_history, metric_history
```

```python
model, loss_hist, metric_hist = train_val(model, params_train)
```
